# making the model

```python
classifier_model = graphlab.logistic_classifier.create(train,target='label',features=['image_array'])
```

```python
classifier_model.evaluate(test)
```

```python
deep_learning_model = graphlab.load_model('http://s3.amazonaws.com/GraphLab-Datasets/deeplearning/imagenet_model_iter45')
```

```python
train['deep_features']=deep_learning_model.extract_features(train)
```

```python
deep_feature_model = graphlab.logistic_classifier.create(train,
                                                        features=['deep_features'],
                                                        target='label')
```

```python
deep_feature_model.evaluate(test)
```

```python
deep_feature_model.predict(test[0:3])
```
